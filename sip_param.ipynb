This scheme will parameterize the number of ice fragments as a function of drop size and temperature. It assumes that the drop is the same temperature as the environment and that the particles are perfectly spherical (the latter being compliant with the expectation of drops spinning as they fall, preserving sphericity).
#====================================================================================================================================================================================
#====================================================================================================================================================================================
#====================================================================================================================================================================================

```python
# Define dependent variables
D_drop = np.arange(0.05,6.01,0.01) # size of drop in mm
temp = np.arange(-35,0.25,0.25) # temperature of drop (and thus of air) in deg C
```

#====================================================================================================================================================================================
#====================================================================================================================================================================================
#====================================================================================================================================================================================
The function gives "N", the number of ice crystal fragments per drop size for the first mode of freezing in which the colliding ice particles are smaller than the size of the drop. The functional form is a Lorentzian whose unknown parameters are determined via observational guidance. These parameters are given as polynomial expressions as a function of drop size. Parameters include Eta (half width of the Lorentzian), Zeta (intensity [i.e. thermal peak] of the Lorentzian), Beta (parameter that includes further temperature dependence for larger drop sizes), and T0 (value of temperature at maximum of Lorentzian). The Lorentzian is assumed to allow a narrow thermal peak and strong wings. Boundary conditions use an interpolating function dependent on temperature (Sigma) and drop size (Xi). Sigma is unity for temperatures < -6 deg C, 0 for temperatures > -3 deg C, and decreases linearly as a multiplicative factor of N between -6 and -3 deg C. Xi is 0 for sizes < 50 microns (D0) and unity above D0 + delta_D = 60 microns. The Lorentzian form and associated thermal peak are associated with observations that the number of drop fragments peaks between -11 and -16 deg C. A proposed theory by Pruppacher and Schlamp (1975; doi: https://doi.org/10.1029/JC080i003p00380) suggests that this peak exists due to higher supercoolings inducing a thicker ice shell (and thus less stress), while at weaker supercoolings, the thinner ice shell more easily cracks to relieve pressure, also decreasing stress on the shell.

```python
#========================================================================================
# Functions
#========================================================================================

# Second degree polynomial function
def polynomial_2nd(coeff1,coeff2,coeff3,dep_var):
    expression = coeff3*(dep_var**2.) + coeff2*(dep_var) + coeff1
    return expression

# Third degree polynomial function
def polynomial_3rd(coeff1,coeff2,coeff3,coeff4,dep_var):
    expression = coeff4*(dep_var**3.) + coeff3*(dep_var**2.) + coeff2*(dep_var) + coeff1
    return expression

#------------------------------------------------------------------------------------------
# Define functions of Lorentzien parameters for equation "N" (Eq. 1 of Phillips et al. [2018])
#------------------------------------------------------------------------------------------
#
# NOTE [1]: All parameter polynomial expressions are represented in Phillips et al. (2018)
# as log(D_drop), such that the dependent variable "X" is given as X = log_10(D_drop). These
# functions are passed D_drop and converted to the logarithmic form within the function.
#
# NOTE [2]: The polynomial expression of Phillips et al. (2018) for the half width (Eta)
# and intensity (Zeta) return log_10(Eta) and log_10(Zeta), respectively. The functions
# here instead return the values of Eta and Zeta in their linear form.
#------------------------------------------------------------------------------------------

# Parameter in expression for N
def beta_func(drop_size):
    drop_size_log = np.log10(drop_size)
    beta_coeff_3 = -0.1839
    beta_coeff_2 = -0.2017
    beta_coeff_1 = -0.0512  
    
    if np.size(drop_size) == 1:
        if drop_size < 0.4:
            beta = 0
        else:
            beta = polynomial_2nd(beta_coeff_1,beta_coeff_2,beta_coeff_3,drop_size_log)
            
    elif np.size(drop_size) > 1:
        beta = np.zeros(np.shape(drop_size))
        tmpid = np.where(drop_size < 0.4)
        beta[tmpid] = 0.
        tmpid = np.where(drop_size >= 0.4)
        beta[tmpid] = polynomial_2nd(beta_coeff_1,beta_coeff_2,beta_coeff_3,drop_size_log[tmpid])
        
    return beta

# Half width of Lorentzian
def eta_func(drop_size):
    eta_coeff_4 = 0.1242
    eta_coeff_3 = -0.2316
    eta_coeff_2 = -0.9874
    eta_coeff_1 = -0.0827
    drop_size_log = np.log10(drop_size)
    eta_log = polynomial_3rd(eta_coeff_1,eta_coeff_2,eta_coeff_3,eta_coeff_4,drop_size_log)
    eta = 10.**(eta_log)
    return eta

    
# Intensity of Lorentzian
def zeta_func(drop_size):
    zeta_coeff_4 = 2.4268
    zeta_coeff_3 = 3.3274
    zeta_coeff_2 = 2.0783
    zeta_coeff_1 = 1.2927 
    drop_size_log = np.log10(drop_size)
    zeta_log = polynomial_3rd(zeta_coeff_1,zeta_coeff_2,zeta_coeff_3,zeta_coeff_4,drop_size_log)
    zeta = 10.**(zeta_log)
    return zeta

# Value of T at maximum of Lorentzian
def T0_func(drop_size):
    T0_coeff_4 = -1.3999
    T0_coeff_3 = -5.3285
    T0_coeff_2 = -3.9847
    T0_coeff_1 = -15.0332
    drop_size_log = np.log10(drop_size)
    T0 = polynomial_3rd(T0_coeff_1,T0_coeff_2,T0_coeff_3,T0_coeff_4,drop_size_log)
    return T0


#------------------------------------------------------------------------------------------
# Define functions of Lorentzien parameters for equation "N_big" (Eq. 3 of Phillips et al. [2018])
#------------------------------------------------------------------------------------------

# Half width of Lorentzian for big drop fragments
def zeta_b_func(drop_size):
    zeta_b_coeff_4 = -0.4651
    zeta_b_coeff_3 = -1.1072
    zeta_b_coeff_2 = -0.4539
    zeta_b_coeff_1 = 0.5317
    drop_size_log = np.log10(drop_size)
    zeta_b = polynomial_3rd(zeta_b_coeff_1,zeta_b_coeff_2,zeta_b_coeff_3,zeta_b_coeff_4,drop_size_log)
    return zeta_b

# Intensity of Lorentzian for big drop fragments
def eta_b_func(drop_size):
    eta_b_coeff_4 = 28.5888
    eta_b_coeff_3 = 49.8504
    eta_b_coeff_2 = 22.4873
    eta_b_coeff_1 = 8.0481
    drop_size_log = np.log10(drop_size)
    eta_b = polynomial_3rd(eta_b_coeff_1,eta_b_coeff_2,eta_b_coeff_3,eta_b_coeff_4,drop_size_log)
    return eta_b

# Value of T at maximum of Lorentzian for big drop fragments
def T0_b_func(drop_size):
    T0_b_coeff_4 = 13.3588
    T0_b_coeff_3 = 15.7432
    T0_b_coeff_2 = -2.6543
    T0_b_coeff_1 = -18.4875
    drop_size_log = np.log10(drop_size)
    T0_b = polynomial_3rd(T0_b_coeff_1,T0_b_coeff_2,T0_b_coeff_3,T0_b_coeff_4,drop_size_log)
    return T0_b
```

```python
#--------------------------------------------------------------------------
#--------------------------------------------------------------------------
# Interpolating functions for onset of fragmentation
#--------------------------------------------------------------------------
#--------------------------------------------------------------------------

# Calculate temperature interpolating function for onset of fragmentation
Tc = -3. # deg C, maximum freezing temperature of drops for fragmentation
#delta_T = np.abs(temp - Tc) # deg C, interval of freezing temperature for onset of fragmentation

omega = np.ones(np.shape(temp))
tmpid = np.where(temp > Tc)
omega[tmpid] = 0
Tc_minus_delta_T = -6.
tmpid = np.where( (temp <= Tc) & (temp >= Tc_minus_delta_T) )
omega[tmpid] = np.linspace(1,0,np.size(tmpid))

# Plot interpolating function for temperature at onset of fragmentation
if False:
#if True:
    fig = plt.figure(figsize=(8,6))
    Fontsize=14
    ax1=fig.add_subplot(111)
    ax1.set_xlim(-35,0)
    ax1.plot(temp,omega,lw=3,c='red')
    ax1.tick_params(labelsize=Fontsize)
    ax1.grid(which='both',lw=1,c='grey')
    ax1.set_xlabel('Temperature [$^{\\circ}$]',fontsize=Fontsize)
    ax1.set_ylabel('Temperature Interpolating Function',fontsize=Fontsize)
    plt.show()
    plt.close()
    
    
# drop size interpolating function
D0 = 50. # microns
D0 = D0*1.e-3 # mm
D0_plus_delta_D = 60 # microns
D0_plus_delta_D = D0_plus_delta_D*1.e-3 # mm

Xi = np.ones(np.shape(D_drop)) 
tmpid = np.where(D_drop < D0)
Xi[tmpid] = 0
tmpid = np.where(D_drop > D0_plus_delta_D)
Xi[tmpid] = 1
tmpid = np.where( (D_drop >= D0) & (D_drop <= D0_plus_delta_D) )
Xi[tmpid] = np.linspace(0,1,np.size(tmpid))

# Plot interpolating function for drop size at onset of fragmentation
if False:
#if True:
    fig = plt.figure(figsize=(8,6))
    Fontsize=14
    ax1=fig.add_subplot(111)
    ax1.plot(D_drop,Xi,lw=3,c='red')
    ax1.tick_params(labelsize=Fontsize)
    ax1.grid(which='both',lw=1,c='grey')
    ax1.set_xlabel('D$_{drop}$ [mm]',fontsize=Fontsize)
    ax1.set_ylabel('Drop size Interpolating Function',fontsize=Fontsize)
    ax1.set_xscale('log')
    plt.show()
    plt.close()
```

```python
# Define function for number of fragments "N" following Phillips et al. (2018), Eq. 1:
# Here, drop_size is a real number and temper is an array. This is used to get N as a function of temperature.
def N_func_temp(drop_size,temper,zeta,eta,T0,beta,omega):
    N = ((zeta*(eta**2.)) / (((temper - T0)**2.) + (eta**2.)) ) + (beta*temper)
    N = N*omega
    return N
```

```python
# Define function for number of fragments "N" following Phillips et al. (2018), Eq. 1:
# Here, drop_size is an array and temper is a real number. This is used to get N as a function of size.
def N_func_size(drop_size,temper,zeta,eta,T0,beta,Xi):
    N = ((zeta*(eta**2.)) / (((temper - T0)**2.) + (eta**2.)) ) + (beta*temper)
    N = N*Xi
    # The above is only valid for D <= 1.6 mm. For larger sizes, we use linear exrapolation up to the largest size.
    tmpid = np.where(drop_size > 1.600000000001)
    tmpid = np.squeeze(tmpid)
    N_at_1_pt_6_mm = N[tmpid[0]-1]
    N[tmpid] = N_at_1_pt_6_mm*(drop_size[tmpid] / 1.6)
    return N
```

```python
# Define function for number of big drop fragments "N_b" following Phillips et al. (2018), Eq. 3:
# Here, drop_size is a real number and temper is an array. This is used to get N as a function of temperature.
def N_b_func_temp(drop_size,temper,zeta_b,eta_b,T0_b,N,omega):
    N_b = (zeta_b*(eta_b**2.)) / (((temper - T0_b)**2.) + (eta_b**2.))
    # Via Eq. 3, N_b = min(N_b,N), so need to also pass in "N" here, the total number of fragments per frozen drop
    N_b = N_b*omega
    for ii in range(len(N_b)):
        N_b[ii] = min(N_b[ii],N[ii])      
    return N_b
```

```python
# Define function for number of big drop fragments "N_b" following Phillips et al. (2018), Eq. 3:
# Here, drop_size is an array and temper is a real number. This is used to get N as a function of size.
def N_b_func_size(drop_size,temper,zeta_b,eta_b,T0_b,N,Xi):
    N_b = (zeta_b*(eta_b**2.)) / (((temper - T0_b)**2.) + (eta_b**2.))
    # Via Eq. 3, N_b = min(N_b,N), so need to also pass in "N" here, the total number of fragments per frozen drop
    N_b = N_b*Xi
    for ii in range(len(N_b)):
        N_b[ii] = min(N_b[ii],N[ii])
    # The above is only valid for D <= 1.6 mm. For larger sizes, we use linear exrapolation up to the largest size.
    tmpid = np.where(drop_size > 1.600000000001)
    tmpid = np.squeeze(tmpid)
    N_b_at_1_pt_6_mm = N_b[tmpid[0]-1]
    N_b[tmpid] = N_b_at_1_pt_6_mm
    return N_b
```

```python
# Define function for number of fragments "N" following Phillips et al. (2018), Eq. 1:
# function is passed an array of both drop size and temperature
def N_func(drop_size,temper,zeta,eta,T0,beta,Xi,omega):
    N = []
    for ii in range(len(temper)):
        tmp_N = ((zeta*(eta**2.)) / (((temper[ii] - T0)**2.) + (eta**2.)) ) + (beta*temper[ii])
        tmp_N = tmp_N*Xi
        N.append(tmp_N*omega[ii])
    N = np.array(N)
    # The above is only valid for D <= 1.6 mm. For larger sizes, we use linear exrapolation up to the largest size.
    tmpid = np.where(drop_size > 1.600000000001)
    tmpid = np.squeeze(tmpid)
    for ii in range(len(N[:,0])):
        N_at_1_pt_6_mm = N[ii,tmpid[0]-1]
        N[ii,tmpid] = N_at_1_pt_6_mm *(drop_size[tmpid] / 1.6)   
    
    return N
```
